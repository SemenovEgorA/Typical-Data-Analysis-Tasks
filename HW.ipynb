```python
from sklearn.datasets import load_boston
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
data = load_boston()
```

```python
df = pd.DataFrame(data["data"], columns = data["feature_names"])
```

```python
df
```

```python
df['y'] = data['target']
```

```python
df
```

```python
import statsmodels.formula.api as smf
```

```python
import statsmodels.api as sm
```

```python
results = smf.ols(formula = 'y ~ LSTAT', data = df).fit()
```

```python
print(results.summary())
```

```python
plt.plot(df['LSTAT'], df["y"], ".", df['LSTAT'], results.predict(), "--" )
```

```python
multResults = results = smf.ols(formula = 'y ~ LSTAT + AGE', data = df).fit()
```

```python
print(multResults.summary())
```

```python
multResults1 = results = smf.ols(formula = 'y ~ CRIM:LSTAT:', data = df).fit()
```

```python
print(multResults1.summary())
```
