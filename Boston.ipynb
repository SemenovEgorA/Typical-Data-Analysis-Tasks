```python
from sklearn.datasets import load_boston
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
data = load_boston()
```

```python
data
```

```python
df = pd.DataFrame(data["data"], columns = data["feature_names"])
```

```python
df
```

```python
df["y"] = data['target']
```

```python
df
```

```python
import statsmodels.formula.api as smf
```

```python
import statsmodels.api as sm
```

```python
results = smf.ols('y ~ LSTAT + CRIM + ZN + np.exp(INDUS) + CHAS + NOX + RM + AGE + DIS + RAD + TAX + PTRATIO + B', data=df).fit()
```

```python
print(results.summary())
```

```python
print(results.predict())
```
